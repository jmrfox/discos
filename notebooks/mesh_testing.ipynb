```python
def process_mesh(filepath_raw, filename_wrapped):
    print(f"Processing {filepath_raw}")
    ms = pymeshlab.MeshSet()
    ms.load_new_mesh(str(filepath_raw))

    # Clean up potential issues
    ms.apply_filter('meshing_remove_duplicate_faces')
    ms.apply_filter('meshing_remove_duplicate_vertices')

    # Fill holes
    # ms.apply_filter('meshing_close_holes')

    # If holes remain or are complex, consider alpha wrapping
    ms.apply_filter('generate_alpha_wrap', alpha_fraction=0.01)

    ms.save_current_mesh(str(filename_wrapped))

ts_numbers = [1, 2, 3, 4, 21, 24, 48, 67, 76]
for n in ts_numbers:
    filepath_raw = data_path(f"mesh/raw/TS{n}_alone.obj")
    filename_wrapped = data_path(f"mesh/processed/TS{n}_wrapped.obj")
    process_mesh(filepath_raw, filename_wrapped)
```
