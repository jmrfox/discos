```python
from discos import *
import numpy as np
import trimesh
import skeletor as sk
import networkx as nx
import matplotlib.pyplot as plt


print("✅ Libraries imported successfully!")
```

```python
mesh = trimesh.load(data_path("mesh/processed/TS2_repaired.obj"))
fixed = sk.pre.fix_mesh(mesh, remove_disconnected=5, inplace=False)
skel = sk.skeletonize.by_wavefront(fixed, waves=1, step_size=1)
print(skel)
# skel.show(mesh=False)
G = skel.get_graph()
nx.draw(G, with_labels=True, arrows=True)
plt.show()
```

```python
from discos import MeshManager, create_cylinder_mesh

mesh = create_cylinder_mesh()
mm = MeshManager(mesh)
# mm.visualize_mesh_3d()

skel = sk.skeletonize.by_wavefront(mesh, waves=1, step_size=1)
print(skel)
# skel.show(mesh=False)

G = skel.get_graph()
nx.draw(G, with_labels=True, arrows=True, pos=skel.swc[["y","z"]].to_numpy())
plt.show()

skel.save_swc(data_path("swc/demos/skeletor_cylinder.swc"))
```
